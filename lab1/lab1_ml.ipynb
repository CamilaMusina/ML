#### 2. Создание функции, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.

```python
def arithmetic_progression(N, first_el, diff):
    diag_el = np.arange(first_el, first_el + N * diff, diff)
    mtr = np.diag(diag_el)
    return mtr
```

```python
N = 10
first_el = 1
diff = 3
res_mtr = arithmetic_progression(N, first_el, diff)
res_mtr
```

#### 16. Построим круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделаем сектор с наибольшим числом квартир выдвинутым.

```python
room_count = data['Комнаты'].value_counts()
room_count
```

```python
plt.figure(figsize=(8, 8))
plt.title('Распределение квартир по количеству комнат')
plt.axis('equal')

max_count_room = room_count.idxmax()
threshold = 20.0

# Создадим категорию "Другие" для маленьких значений
room_count_filtered = room_count[room_count.values >= threshold]
small_values = room_count[room_count.values < threshold]
room_count_optimized = room_count_filtered.copy()
room_count_optimized['Другие'] = small_values.sum()

explode = [0.05 if room == max_count_room else 0 for room in room_count_optimized.index]
labels = [f'{room} ({count:.1f})' for room, count in zip(room_count_optimized.index, room_count_optimized)]

plt.pie(room_count_optimized, labels=labels, autopct='%1.1f%%', startangle=0, counterclock=False, explode=explode)
legend_labels = [f'{room}: {count:.1f}' for room, count in zip(room_count_optimized.index, room_count_optimized)]
plt.legend(legend_labels, title="Количество комнат")
legend = plt.legend(legend_labels, title="Количество комнат", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

#### 21. При помощи сетки графиков PairGrid визуализируем попарные отношения признаков Rooms, Square, HouseFloor, HouseYear, Price следующим образом: на диагонали - гистограммы, под диагональю - ядерные оценки плотности, над диагональю - диаграммы рассеяния. 

```python
def plot_pairgrid(data):
    gr = sns.PairGrid(data, vars=["Комнаты", "Площадь", "Этажность Дома", "Год Постройки", "Стоимость"])
    gr.map_diag(sns.histplot, bins=20)
    gr.map_upper(sns.scatterplot)
    gr.map_lower(sns.kdeplot, bw_method=0.5)
    gr.fig.suptitle("PairGrid Plot of Rooms, Square, HouseFloor, HouseYear, Price")
    plt.show()
```

```python
num_processes = 4
data_ch = np.array_split(data.sample(n=2500), num_processes)
combined_data = pd.concat(data_ch)
plot_pairgrid(combined_data)
```
